**NETFLIX INTERACTIVE RECOMMMENDATION SYSTEM **

The Netflix Interactive Recommendation System is used to provide personalized content recommendations based on user preferences, viewing history, and trends.
 It helps users discover movies and shows that match their interests using machine learning algorithms.

✔ **View Top Genres** – Identify the most-watched genres on Netflix.
✔ **Analyze Specific Year Data** – Filter content released in 2024 (or any other year).
✔** Visualize Trends** – Use charts (bar, pie, line) to analyze Netflix content.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Netflix Data new.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Release Year'].max() #to identify most recent year present in the dataset
```

```python
df[df['Release Year']==2024]["Main Genre"].value_counts() #how many times each genre appears in the dataset for 2024
```

```python
df2024=df[df['Release Year']==2024]
```

```python
df2024
```

```python
dfyr=df2024['Main Genre'].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.pie(dfyr.values,labels=dfyr.index,autopct="%1.1f%%")
plt.show()
```

**BAR CHART** (TOP 10 MOST GENRE ON NETFLIX)

```python
plt.figure(figsize=(12,6))
genre_counts= df["Main Genre"].value_counts().head(10)
sns.barplot(x=genre_counts.values,y=genre_counts.index,palette="viridis")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.title("Top 10 Genre on Netflix")
plt.show()
```

**LINE CHART** (TREND OF NETFLIX CONTENT PRODUCTION OVER THE YEARS)

```python
plt.figure(figsize=(12,6))
Year_counts=df['Release Year'].value_counts()
sns.lineplot(x=Year_counts.index,y=Year_counts.values,marker="o",linestyle="-",color="deeppink")
plt.xlabel("Release Year")
plt.ylabel("No. of Titles")
plt.title("TREND OF NETFLIX CONTENT PRODUCTION OVER THE YEARS")
plt.show()
```
